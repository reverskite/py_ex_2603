```python
# 파이썬 유니코드 / 인코딩 핵심 정리

# 문자열 컴퓨터 내부에서 숫자(유니코드)로 저장된다.
# -내부적으로는 각 글자를 숫자로 바꿔 저장한다.
# - 그 숫자 체계를 유니코드(uinicode)라고 한다
text = '파이썬'

print(text) # 문자열 출력
print(len(text)) # 문자열 길이

# 문자 숫자 변환
# String : 문자열
# char : 문자
print(ord('가')) # 문자를 유니코드 숫자로 변환
print(chr(44032)) # 숫자를 다시 문자로 변환

# 3. 문자열을 바이크로 변환 (encode)
# - encode : 문자열을 바이트(byte)로 바꾸는 작업
# 파일 저장이나 네트워크 전송 시 사용됨
encoded = text.encode('utf-8')
print(encoded)
print(type(encoded))

# 바이트를 다시 문자열로 변환 (decode)
decoded = encoded.decode('utf-8')
print(decoded)
print(type(decoded))

# 5. 인코딩이 맞지 않으면 오류 발생
try: 
    wrong = encoded.decode('cp949')
    print(wrong)

except FileNotFoundError:
    print("파일을 찾을 수 없습니다.")

# unicode : 문자 -> 번호
# UTF-8 : 번호 -> 바이트
# encode : 문자열 -> bytes
# decode : bytes -> 문자열
```
