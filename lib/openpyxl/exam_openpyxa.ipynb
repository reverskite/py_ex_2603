```python
import openpyxl
from openpyxl import Workbook, load_workbook

# 1. 새 엑셀 파일 생성 및 저장
w1 = Workbook()
w1.save('text.xlsx')

# 2. 기존 파일 불러오기
try:
    w2 = load_workbook('text.xlsx')
    ws2 = w2.active

    # 3. 셀에 값 입력 (A1)
    ws2['A1'] = 'hello'

    # 4. C열 3행에 값 입력
    # 'good'은 문자열이므로 반드시 따옴표('')로 감싸야 합니다.
    ws2["C3"] = "good" 

    # 5. 범위 내 셀에 값 채우기 (D5:F8)
    a = 'python'
    # 'for row in ... :' 형태로 작성해야 하며 끝에 콜론(:)이 필수입니다.
    for row in ws2['D5':'F8']: 
        for cell in row:
            cell.value = a

    # 6. 최종 저장
    # 파일명 오타 수정: .slsx -> .xlsx
    w2.save('text.xlsx') 
    print("파일 수정 및 저장 완료!")
    
except FileNotFoundError:
    print("파일을 찾을 수 없습니다.")
    
    
```

```python
from openpyxl import *

# === 딕셔너리로 값 생성 후 엑셀에 넣기 ===

item1 = {'품목': '바나나', 
         '가격': '3000', 
         '수량': 1, 
         '날짜': '2026년 3월 13일'
}

item2 = {'품목': '커피',
         '가격': '2500',
         '수량': '3',
         '날짜': '2026년 3월 15일'
}

item3 = {'품목': '과자',
         '가격': '1500',
         '수량': '4',
         '날짜': '2026년 3월 15일'
}

wb = Workbook()
ws = wb.active

ws['A1'] = '품목' # A열 1행 : 품목
ws['B1'] = '가격' # B열 1행 : 가격
ws['C1'] = '수량' # C열 1행 : 수량
ws['D1'] = '날짜' # D열 1행 : 날짜

items = [item1, item2, item3]

i = 2
for item in items:
    ws['A' + str(i)] = item['품목']
    ws['B' + str(i)] = item['가격']
    ws['C' + str(i)] = item['수량']
    ws['D' + str(i)] = item['날짜']
    i += 1
    
wb.save('test2.slsx')
```
