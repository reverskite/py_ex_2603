```python
# 문제 : 실행되는 출력문에는 `참` 그렇지 않으면 `거짓` 이라고 적어주세요.
# 조건 : `참1`, `참2`, `거짓1` 이런 순으로 적어주세요.

# 개념시작
if ( True ):
  print("참")
    
if ( False ):
  print("거짓")
# 개념끝

# 문제시작
a = 10

# `==` => 같다.
if ( a == 10 ):
  print("참1")

# `!=` => 같지 않다.
if ( a != 10 ):
  print("참2")

if ( a > 10 ):
  print("거짓1")

if ( a >= 10 ):
  print("참3")

b = 10

if ( a == b ):
  print("참4")

# c 에는 true or false 이 담긴다.
c = a != b # c = False

if ( c ):
  print("거짓")

if ( c == False ):
  print("참")

# `not` => 반전
if ( not c ):
  print("참")

if ( not(not c) ):
  print("거짓")

d = True

if ( c != d ):
  print("참")

if ( 10 > 5 and -6 > -10 and 10 != 10 ):
  print("거짓4")

if ( 10 < 5 and -6 < -10 and 10 != 10 or 10 > 5 ):
  print("참7")
# 문제끝
```

```python
s = "안녕하세요."

print("s의 길이 출력")
print(len(s))

# 문자열[starr:end(end - 1):step]
print("s를 이루는 문장 중에서 앞에 거, 2개 출력")
print(s[0:2])

print("s를 이루는 문장 중에서 문자 하나 건너 뛰고 1개 출력")
print(s[1:2])

print("s를 이루는 문장 중에서 앞에 문자 4개 출력, 첫 인덱스 생력")
print(s[:4])

print("s를 이루는 문장 중에서 앞에 문자 3개 건너 뛰고, 나머지 전부 출력, len")
print(s[2:len(s)])

print("== 예제 2 ==")
s = "안녕하세요"
   # 0 1 2 3 4 5

print("s를 이루는 문자를 하나씩 출력, while")
i = 0
while i < len(s):
    print(s[i])
    i += 1
    
# string(문자열/문장), char(문자 하나)
print("s를 이루는 문자를 하나씩 출력, for")
for c in s:
    print(c)
```

```python
# 클래스 -> 설계도
class person:
    def Hello(self):
        print("hi")


jo = person() # 변수 = 객체 jo는 person의 인스턴스
jo.Hello() # 인스턴스 메서드

# 빈 클래스 만드는 방법
class person:
    def Hello(self):
        print("Hello")
        
    def hi(self):
        self.Hello()
        
jo = person()


```

```python
# 클래스 속성

class person:
    folder = [] # 클래스 속성(공공제) -> 클래스에 속한다. 모든 인스턴스에서 접근이 가능하다.
    
     
    def add_files(self, file):
        self.folder.append(file)
        
        
jin = person()
jin.add_files('txt')

paul = person()
paul.add_files('txt')

```

```python
# 비공개 클래스 속성

class person:
    __weight = 78
    
    def print_weight(self):
        print(person._weight)
        
jin = person()
jin.person_weight()
```
